```python
# imports

import sqlite3
```

```python
# speech loading function
def load_speeches_by_pope(db_path="../../data/vatican_speeches_test.db"):
    conn = sqlite3.connect(db_path)
    conn.row_factory = sqlite3.Row
    cur = conn.cursor()

    # Get all popes
    cur.execute("SELECT pope_name FROM popes")
    popes = [row["pope_name"] for row in cur.fetchall()]

    result = {}

    for pope in popes:
        cur.execute(
            """
            SELECT title, text
            FROM speeches
            WHERE pope_name = ?
            ORDER BY date
            """,
            (pope,),
        )
        speeches = [
            {"title": row["title"], "text": row["text"]} for row in cur.fetchall()
        ]

        result[pope] = speeches

    conn.close()
    return result
```

```python
pope_data = load_speeches_by_pope()
```

```python
for pope, speeches in pope_data.items():
    print(pope, len(speeches))
```
